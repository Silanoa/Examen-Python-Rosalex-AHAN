```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1,20)
y = x**5 + 0.5
plt.subplots(figsize=(4, 7))
plt.plot(x,y,color='r')
plt.plot([0,0],[0,1],color='r')
plt.plot([1,1],[0,1],color='r')
plt.plot([0,1],[0,0],color='r')
plt.plot([0,1],[1,1],color='r')
plt.grid()
plt.show()
```
